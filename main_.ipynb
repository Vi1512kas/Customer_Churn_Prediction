### Encoding Categorical Data....


```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
lb = LabelEncoder()
data['Gender'] = lb.fit_transform(data['Gender'])
```

```python
data
```

```python
data
```

```python
X = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
trans = ColumnTransformer(
    transformers = [
        ('ohe',OneHotEncoder(drop='first'),["Geography"])
    ],
    remainder='passthrough'
)
# trans.set_output(transform="default")
X_train = trans.fit_transform(X_train)
X_test = trans.transform(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import pickle
with open('scaler.pkl','wb') as f1:
    pickle.dump(scaler,f1)
    
```

```python
with open('tranformer.pkl','wb') as f2:
    pickle.dump(trans,f2)
```

```python
with open('label.pkl','wb') as f3:
    pickle.dump(lb,f3)
```

```python
### ANN Implementation...
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from tensorflow.keras.models import Sequential
import datetime
```

```python
model = Sequential(
    [
        Dense(64,activation='relu',input_shape=(X_train.shape[1],)), ## first layer with input interaction
        Dense(32,activation='relu'), ## HL2
        Dense(1,activation='sigmoid') # Output Layer
    ]
)
```

```python
model.summary()
```

```python
import tensorflow
```

```python
### if we want our learning rate for different loss.
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
### for fixed learning rate.
### Compiling the model.
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model
```

```python
### setting up the tensorboard.

### to create logs for visualisation.
log_dir = "logs/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# set up early Stopping CallBack.
## to stop at some extent.
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    X_train_scaled,y_train,validation_data=(X_test_scaled,y_test),epochs=100,
    callbacks = [tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
